```
import pandas as pd
import jebas as jb
import seaborn as sns
import matplotlib.pyplot as plt
```

# Dados

```
franquias = pd.read_excel(
    'dados_mercado_franquia_2017_2020.xlsx',
    sheet_name='trimestre_python'                     
)
```

```
novo_periodo = {}
contador=1
for periodo in franquias['periodo'].unique():
    dict_temp = {periodo:contador}
    novo_periodo.update(dict_temp)
    contador +=1

franquias['novo_periodo'] = franquias['periodo'].map(novo_periodo)
```

# Regressão e Previsão

# Statsmodels

## Teste 1 - Base completa

## Preparação dos dados

```
import statsmodels.api as sm
```

```
Y = franquia_selecao['faturamento']
X = sm.add_constant(franquia_selecao['novo_periodo'])
```

## Regressão

```
resultado_regressao = sm.OLS(Y, X, missing='drop').fit()
```

```
resultado_regressao.params
```

```
resultado_regressao.conf_int()
```

```
franquia_selecao['y_previsao'] = resultado_regressao.predict()
franquia_selecao['residuo'] = resultado_regressao.resid
franquia_selecao
```

```
grafico7_a = sns.lineplot(
    data=franquia_selecao,             
    x=franquia_selecao['novo_periodo'],             
    y=franquia_selecao['faturamento']             
)
grafico7_b = sns.lineplot(
    data=franquia_selecao,             
    x=franquia_selecao['novo_periodo'],             
    y=franquia_selecao['y_previsao']             
)
grafico7_a.figure.set_size_inches(20,6)
grafico7_a.set_title('Gráfico 7 - Checagem Regressão Linear', color='r', fontsize=22)
grafico7_a.set_xlabel('Faturamento', color='r', fontsize=18)
grafico7_a.set_ylabel('Frequencia', color='r', fontsize=18)
grafico7_a.axis(ymin=0, ymax=franquia_selecao['faturamento'].max() * 1.2)
plt.show()
```
